## Add the average violation records of the past 1 year of inspections to iprove the performance

```python
import pandas as pd
import numpy as np
```

```python
# Load data and print stats about the dataset
data = pd.read_csv('../dataset/iy12th2.csv')
data = data.drop(['Unnamed: 0'], axis=1)
print("n_of_inspections", data.shape[0])
print("\nn_of_unique_restaurts: ", len(data['ID'].unique()))
print("\nColumn names \n", data.columns.get_values())
data.head(3)
```

```python
# Add duplicated columns of inspections
dm = data.merge(data[['ID', 'Inspect Dt', 'Low', 'Medium', 'High']], on='ID')
dm.shape
```

```python
# Change the data type into datetime
dm['Inspect Dt_x'] = pd.to_datetime(dm['Inspect Dt_x'])
dm['Inspect Dt_y'] = pd.to_datetime(dm['Inspect Dt_y'])
```

```python
# Example result
dmt.loc[(dmt['ID']==50755)]
# We will select only past one year data from this
```

```python
# Calculate day difference between inspection_dt_x and inspection_dt_y
def get_day_diff(row):
    return (row['Inspect Dt_x'] - row['Inspect Dt_y']).days

dm['insp_dt_diff'] = dm.apply(get_day_diff, axis=1)

# Include only past 1 year of inspection data
dmt = dm.loc[(dm['insp_dt_diff']<=365)&(dm['insp_dt_diff']>0)]
```

```python
dmt.columns.get_values()
```

```python
print("n_of_inspections: ", dmt.shape[0])
print("\nn_of_unique_restaurants: ", len(dmt['ID'].unique()))
```

```python
# Make the past violation data into average of violations
dmtg = dmt.groupby(['Encounter', 'ID', 'Facility Name', 'Bus St Date', 'Description',
       'Inspect Dt_x', 'Low_x', 'Medium_x', 'High_x', 'business_id',
       'address', 'price', 'cuisine_Asian Fusion', 'cuisine_Vegetarian',
       'cuisine_Chinese', 'cuisine_American (Traditional)',
       'cuisine_Italian', 'cuisine_Mexican', 'cuisine_Thai',
       'cuisine_others', 'cuisine_American (New)', 'cuisine_Indian',
       'cuisine_Japanese', 'cuisine_Greek', 'cuisine_Mediterranean',
       'hyg_0', 'hyg_1', 'hyg_2', 'hyg_3', 'hyg_4', 'hyg_5', 'unh_0',
       'unh_1', 'unh_2', 'unh_3', 'stars_y', 'count'],
        as_index=False)['Low_y', 'Medium_y', 'High_y'].mean()
```
